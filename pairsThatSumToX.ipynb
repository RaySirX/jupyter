```python
import random
```

```python
arr = sorted(random.sample(range(1,1000),500))
```

```python
print(arr)
print(len(arr))
```

```python
def pairsThatSumToX(a):
   for j in sorted(a):
       for k in sorted(a, reverse=True):
          if j > k:
            yield (j,k)
```

```python
print([y for y in pairsThatSumToX(arr) if y[0]+y[1] == 81])
```
