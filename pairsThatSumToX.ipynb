```python
import random
```

```python
arr = sorted(random.sample(range(1,1000),500))
```

```python
print(arr)
print(len(arr))
```

```python
def uniquePairs(a):
   for j in sorted(a, reverse=True):
       for k in sorted(a):
          if j > k:
            yield (j,k)
```

```python
print(len([y for y in uniquePairs(arr)]))
```

```python
print([y for y in uniquePairs(arr)])
```

```python
print([y for y in uniquePairs(arr) if y[0]+y[1] == 81])
```
